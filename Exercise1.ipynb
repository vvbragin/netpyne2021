```python
from neuron import h
from neuron.units import µm, mV, ms
import numpy as np
import matplotlib.pyplot as plt
import plotly
h.load_file("stdrun.hoc")

axon = h.Section(name="axon")
axon.diam = 1 * µm
axon.L = 3000 * µm
axon.nseg = 1001

h.hh.insert(axon)

ic = h.IClamp(axon(0))
ic.delay = 1 * ms
ic.amp = 1 # nA
ic.dur = 0.1 * ms

rvp = h.RangeVarPlot('v', axon(0), axon(1))

h.finitialize(-65 * mV)

plot = rvp.plot(plotly, name="t=0")

for tStop in np.arange(0.5, 5.5, 0.5):
  h.continuerun(tStop)
  plot = rvp.plot(plot, name=f"t={tStop}")

plot = plot.update_layout({"xaxis_title": "position (µm)", "yaxis_title": "membrane potential (mV)"})

plot.show()

```

```python
# visualize with color-coding
ps = h.PlotShape(False)
ps.variable("v")
ps.scale(-65, 30)
ps.plot(plotly).show()
```
