## Sodium and Potassium current responses to being voltage clumped

```python
from neuron import h
from neuron.units import mV, ms, um
import plotly.graph_objects as go
h.load_file('stdrun.hoc')
```

```python

soma = h.Section(name="soma")
soma.L = soma.diam = 10 * um

h.hh.insert(soma)

vclamp = h.SEClamp(soma(0.5))
vclamp.amp1 = -65 * mV; vclamp.dur1 = 5 * ms
vclamp.amp2 = 0 * mV; vclamp.dur2 = 20 * ms
vclamp.amp3 = -65 * mV; vclamp.dur3 = 10000 * ms

t = h.Vector().record(h._ref_t)
ina = h.Vector().record(soma(0.5)._ref_ina)
ik = h.Vector().record(soma(0.5)._ref_ik)

h.finitialize(-65 * mV)
h.continuerun(35 * ms)

fig = go.Figure()
fig.add_trace(go.Scatter(x=t, y=ina, name="ina"))
fig.add_trace(go.Scatter(x=t, y=ik, name="ik"))
```

```python
fig = go.Figure()

for v in range(-80, 40, 10):
  vclamp.amp2 = v
  h.finitialize(-65 * mV)
  h.continuerun(30)
  fig.add_trace(go.Scatter(x=t, y=ik, name=f"v={v}", line={"width": 4}))

fig.update_layout(dict(xaxis_title="t (ms)", yaxis_title="potassium current (mA/cm**2)",))

fig.show()
```
